```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import pearsonr, spearmanr

from asa.plot_methods import plot_trend
```

```python
x = np.random.uniform(0, 1.2, 1000)
y = 1.5 * x

x_obs = np.random.normal(x, x/3)
y_obs = np.random.normal(y, y/3)
```

```python
good = (x_obs > 0) & (y_obs > 0)
```

```python
plt.scatter(x_obs[good], y_obs[good] / x_obs[good])
plot_trend(x_obs[good],
           y_obs[good] / x_obs[good],
           fkind='errorbar',
           bins=10,
           errorbar_kwargs={'color': 'red'})
plt.ylim(0, 5)
```

```python
pearsonr(x_obs[good], y_obs[good] / x_obs[good])
```

```python
spearmanr(x_obs[good], y_obs[good] / x_obs[good])
```
