```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import powerlaw, laplace
from asa.plot_methods import plot_contour, plot_trend
```

```python
x = powerlaw.rvs(a=0.3, size=100000)
plt.hist(x)
```

```python
y = laplace.rvs(size=100000)
plt.hist(y)
```

```python
plot_contour(x, y, range=[[0, 1], [-3, 3]], bins=30, kde_smooth=True)
```

```python
plot_contour(x, y, range=[[0, 1], [-3, 3]], bins=30, kde_smooth=True)
plot_trend(x,
           y,
           plot_kwargs={'color': 'blue'},
           plot_scatter_kwargs={'color': 'blue'},
           ifscatter=True)
```
