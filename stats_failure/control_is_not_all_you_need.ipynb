```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

from asa.plot_methods import plot_contour, plot_trend
```

```python
N = 100000

class_type = np.random.randint(0, 2, size=N)

sport_skill = np.random.normal(5, 2, size=N)
music_skill = np.random.normal(5, 2, size=N)

total_score = np.zeros(N)
total_score[class_type == 0] = sport_skill[class_type == 0] + music_skill[class_type == 0] + np.random.normal(0, 2, size=(class_type == 0).sum())
total_score[class_type == 1] = sport_skill[class_type == 1] + music_skill[class_type == 1] + np.random.normal(0, 4, size=(class_type == 1).sum())
```

```python
plt.hist(total_score[class_type == 0], alpha=0.5, label='class 0')
plt.hist(total_score[class_type == 1], alpha=0.5, label='class 1')
plt.xlabel('total score')
plt.legend()
```

```python
in_score_bin = (total_score > 8) & (total_score < 9)
```

```python
plot_contour(sport_skill[in_score_bin & (class_type == 0)], music_skill[in_score_bin & (class_type == 0)], color='blue', plot_density=False, kde_smooth=True)
plot_contour(sport_skill[in_score_bin & (class_type == 1)], music_skill[in_score_bin & (class_type == 1)], color='red', plot_density=False, kde_smooth=True)

plot_trend(sport_skill[in_score_bin & (class_type == 0)], music_skill[in_score_bin & (class_type == 0)], plot_kwargs={'color': 'darkblue', 'label': 'class 0'}, ifscatter=True)
plot_trend(sport_skill[in_score_bin & (class_type == 1)], music_skill[in_score_bin & (class_type == 1)], plot_kwargs={'color': 'darkred', 'label': 'class 1'}, ifscatter=True)

plt.legend()

plt.xlabel('sport skill')
plt.ylabel('music skill')

# pearson'r
print('pearsonr (class 0): ', pearsonr(sport_skill[in_score_bin & (class_type == 0)], music_skill[in_score_bin & (class_type == 0)]))

print('pearsonr (class 1): ', pearsonr(sport_skill[in_score_bin & (class_type == 1)], music_skill[in_score_bin & (class_type == 1)]))
```
